```python
import time
import  matplotlib.pyplot as plt
```

```python
x =[]
y =[]
def fib1(n):
    if n <= 0:
        return 0
    if n==1:
        return 1
    return fib1(n - 1) + fib1(n - 2)  
```

```python
x =[]
y=[]
def fib2(n):
    lst = [0] * (n + 1)
    lst[0] =0
    lst[1] = 1
    for i in range(2, n + 1):
        lst[i] = lst[i - 1] + lst[i - 2]
    return lst[n]
```

```python
for i in range (5,45,5):
    x.append(i)
    a=time.perf_counter_ns()
    fib2(i)
    b = (time.perf_counter_ns())
    c = b-a
    y.append(c)
```

```python
plt.plot(x, y,'o')
plt.title('Time of fib1(n)')
plt.xlabel('n')
plt.ylabel('Time(s)')
plt.show() 
```
