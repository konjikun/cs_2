```python
import matplotlib.pyplot as plt
```

```python
origin = 100000
inc = origin * 0.02

xs = [i for i in range(1,51)]
ys =[ origin + inc*i for i in range(1,51)]

xs2 = [i for i in range(1,51)]
ys2 = []
for i in range(1,51):
    ys2.append(origin +origin * 0.02)
    origin = origin + origin *0.02

plt.plot(xs2,ys2,label = 'Compound')
plt.plot(xs, ys, label ='Simple')
plt.plot()
plt.xlabel("Year")
plt.ylabel("Amount")
plt.title("Simple and Compound interest")
plt.legend()
```
