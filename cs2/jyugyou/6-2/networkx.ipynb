```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G1 = nx.Graph()

G1.add_node(1)
G1.add_node(2)
G1.add_node(3)

G1.add_edge(1,2)
G1.add_edge(2,3)
G1.add_edge(1,3)
```

```python
nx.draw(G1, with_labels=True)
```

```python
G3 = nx.DiGraph()

G3.add_node('a')
G3.add_node('b')
G3.add_node('c')
G3.add_node('d')
G3.add_node('e')
G3.add_node('f')

G3.add_edge('a', 'b')
G3.add_edge('a', 'c')
G3.add_edge('a', 'd')
G3.add_edge('b', 'a')
G3.add_edge('c', 'f')
G3.add_edge('d', 'e')
G3.add_edge('e', 'b')
G3.add_edge('e', 'd')
G3.add_edge('f', 'e')
```

```python
nx.draw(G3, with_labels=True)
```

```python
list(nx.dfs_edges(G3, 'a'))
```

### ここから自分の実験

```python
bill =nx.DiGraph()

for handred in range(1,101):
    bill.add_node(handred)

for oneDiff in range(1,100):
    bill.add_edge(oneDiff,oneDiff+1)
    bill.add_edge(oneDiff+1,oneDiff)
    ##50階以上の時に100の条件を超えてしまうため
    if oneDiff <51:
        bill.add_edge(oneDiff,oneDiff*2)
    
    if oneDiff %3==0:
        bill.add_edge(oneDiff,oneDiff/3)

        
    
        
    
```

```python
nx.draw(bill,with_labels=True)
```

```python
list(nx.shortest_path(bill,33,77))
```
