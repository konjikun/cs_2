```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G1 = nx.Graph()

G1.add_node(1)
G1.add_node(2)
G1.add_node(3)

G1.add_edge(1,2)
G1.add_edge(2,3)
G1.add_edge(1,3)
```

```python
nx.draw(G1, with_labels=True)
```

```python
G3 = nx.DiGraph()

G3.add_node('a')
G3.add_node('b')
G3.add_node('c')
G3.add_node('d')
G3.add_node('e')
G3.add_node('f')

G3.add_edge('a', 'b')
G3.add_edge('a', 'c')
G3.add_edge('a', 'd')
G3.add_edge('b', 'a')
G3.add_edge('c', 'f')
G3.add_edge('d', 'e')
G3.add_edge('e', 'b')
G3.add_edge('e', 'd')
G3.add_edge('f', 'e')
```

```python
nx.draw(G3, with_labels=True)
```

```python
list(nx.dfs_edges(G3, 'a'))
```
