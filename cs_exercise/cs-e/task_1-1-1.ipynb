正の整数からなるリストxsを引数に取り、xsの要素の中で3の倍数の個数を返す関数vount3(xs)を定義せよ。

```python
def count3(xs):
    a=[i for i in xs if i%3==0]
    result=len(a)
    return result
```

```python
count3([123,456,789])
```

単語のリストwordsを引数にとり、空白を挟んでつないだ文字列を返す関数sentence(words)を定義せよ。

```python
def sentence(words):
    return ' '.join(words)
```

```python
sentence(['This','is','INIAD'])
```
