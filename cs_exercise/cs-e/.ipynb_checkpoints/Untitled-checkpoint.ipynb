### 出欠問題

```python
def sum(n):
    if n <= 1:
        return 1
    return sum(n-1) + n 

print(sum(4))
```

### グラフの作成

```python
import networkx as nx
import matplotlib as pyplot
```

```python
G1 = nx.Graph()  ## 有向グラフはDigGraph()
```

```python
for i in range(1,4):
    G1.add_node(i)
G1.add_edge(1,2)
G1.add_edge(2,3)
G1.add_edge(1,3)
```

```python
nx.draw(G1, with_labels =True)
```
