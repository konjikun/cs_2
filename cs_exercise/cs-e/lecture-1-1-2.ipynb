##1-1-2 Matplotlib でグラフを描画する

```python
import matplotlib.pyplot as plt
import math
```

```python
xs = [0,1,2,3]
ys = [0,3,2,3]
ys2=[3,2,1,0]

plt.plot(xs,ys,'o', label = 'sample')
plt.plot(xs,ys2,'x', label = 'sample2')

plt.title('Sample graph')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```
