##　課題１


以下の5つの関数について、'$1\leqq x\leqq8$'の区間のグラフを重ねて描画する  
・'$y = x$'  
・'$y = x^2$'  
・'$y = 2^x$'  
・'$y = \log ^x$'  
・'$y = x \log x$'


```python
import  matplotlib.pyplot as plt
import math

y1=[]
y2=[]
y3=[]
y4=[]
y5=[]
x1=[]
x2=[]
x3=[]
x4=[]
x5=[]

for x in range (100,801):
    x*=0.01
    y1.append(x)
    y2.append(x**2)
    y3.append(2**x)
    y4.append(math.log(x))
    y5.append(x*math.log(x))
    x1.append(x)
    x2.append(x)
    x3.append(x)
    x4.append(x)
    x5.append(x)

plt.xlim(1,10)
plt.ylim(0,250)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphs of several functions')
plt.plot(x1,y1,label='$y = x$')
plt.plot(x2, y2,label='$y = x^2$')
plt.plot(x3, y3,label='$y = 2^x$')
plt.plot(x4, y4,label='$y = \log ^x$')
plt.plot(x5, y5,label='$y = x \log x$')
plt.legend()
plt.show()
```

### 双曲線のグラフの実験

```python
import numpy as np
from matplotlib import pyplot as plt


def MakeGraph(a:int, p:int, q:int) -> None:

    ########## 分数関数をグラフに図示 ##########
    """
    分数関数のxを漸近線 x = p について、
    0.1間隔でそれぞれ
    双曲線①：(p + 0.1)     < x < (p + 0.1) + 5
    双曲線②：(p - 0.1) - 5 < x < (p - 0.1)
    の範囲を指定
    """

    # 双曲線共通の設定（xの値を0.1の間隔で設定するための変数）
    xinterval = 0.1

    # 双曲線①
    xmin_1 = p + 0.1
    xmax_1 = xmin_1 + 5
    x1 = np.arange(xmin_1, xmax_1, xinterval)

    # yを算出
    y1 = a/(x1 - p) + q

    # グラフを作成
    plt.plot(x1,y1,color = "blue")

    # 双曲線②
    xmax_2 = p - 0.1
    xmin_2 = xmax_2 - 5
    x2 = np.arange(xmin_2, xmax_2, xinterval)

    y2 = a/(x2 - p) + q

    # グラフを作成
    plt.plot(x2,y2,color = "blue")

    ########################################


    # グラフを表示
    plt.show()


# メイン処理
if __name__ == '__main__':
    # y = 1/(x - 2) + 1
    a = 1
    p = 2
    q = 1
    MakeGraph(a, p, q)
```
